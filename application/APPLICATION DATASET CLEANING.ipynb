```python
df['x_percentage'] = df['x_percentage'].astype(str).str.replace('%', '').str.strip()
df['x_percentage'] = pd.to_numeric(df['x_percentage'], errors='coerce')

df['bachelor_percentage'] = df['bachelor_percentage'].astype(str).str.replace('%', '').str.strip()
df['bachelor_percentage'] = pd.to_numeric(df['bachelor_percentage'], errors='coerce')
```

```python
df['x_percentage'].fillna(df['x_percentage'].median(), inplace=True)
df['bachelor_percentage'].fillna(df['bachelor_percentage'].median(), inplace=True)
```

```python
percentage_columns = ['x_percentage', 'xii_percentage', 'bachelor_percentage']

for col in percentage_columns:
    df[col] = df[col].astype(str).str.replace('%', '').str.strip()

    df[col] = pd.to_numeric(df[col], errors='coerce')

    df[col].fillna(df[col].median(), inplace=True)
```
