**OBJECTIVE — To extract core keywords from applicant SOPs to gain insights into common academic goals, technical skills, and research interests expressed across applications.**


```python


import pandas as pd
import string   #string - to remove punctuation
from collections import Counter # Counter - to count word frequencies
```

```python
df = pd.read_csv(r"C:\Users\Admin\Downloads\Cleaned_Applications.csv")

```

```python
df.head()
```

```python
sop_data = df["statement_of_purpose"].dropna()
```

```python
stopwords = set([
    "i", "me", "my", "myself", "we", "our", "you", "your", "he", "she", "it", "they",
    "is", "are", "was", "were", "be", "been", "have", "has", "had", "do", "does",
    "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
    "of", "at", "by", "for", "with", "about", "into", "through", "during", "to", "from",
    "in", "on", "that", "this", "so", "can", "will", "just", "not", "am", "would", "could", "should"
])
#This is a manually defined list of common English words(stopwords) that are not useful for keyword analysis.
#STOPWORDS ARE USUALLY OMMITED FROM THE ANALYSIS


```

```python
all_words = []

for text in sop_data:
    text = str(text).lower()  # Convert to lowercase
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    words = text.split()  # Tokenize
    words = [word for word in words if word not in stopwords]  # Remove stopwords
    all_words.extend(words)

#Each SOP is lowercased, punctuation removed, then split into words. Stopwords are removed, and the cleaned words are collected.
```

```python
word_freq = Counter(all_words)
#Counter to count how many times each word appears across all SOPs.


```

```python
top_keywords = word_freq.most_common(20)
for word, count in top_keywords:
    print(f"{word}: {count}")
# Display the top 20 most frequent keywords with their counts.
# We can change 20 to some other number and it will display accordingly 

```
