#Data cleaning

```python
df.drop(columns=['dob', 'gender', 'roll_no', 'degree_name'], inplace=True)
```

```python
df = df[~df['name'].str.lower().str.contains("test", na=False)]
```

```python
# Identify percentage columns
percent_cols = [col for col in df.columns if "percentage" in col]

# Convert % strings to float numbers
for col in percent_cols:
    df[col] = df[col].str.replace('%', '').astype(float)
```

```python
df['start_time'] = pd.to_datetime(df['start_time'], errors='coerce')
df['submit_time'] = pd.to_datetime(df['submit_time'], errors='coerce')
```

```python
df['calculated_duration_minutes'] = (df['submit_time'] - df['start_time']).dt.total_seconds() / 60
df['calculated_duration_minutes']
```

```python
# Who paid the token
df['is_token_paid'] = df['token_paid'].notna()
df['is_token_paid']
# Who switched tabs (more than 0)
df['tabs_switched_flag'] = df['tabswitched'].apply(lambda x: 'Yes' if x > 0 else 'No')
df['tabs_switched_flag']
```

```python
# Check which categorical (object) columns have missing values
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols].isnull().sum().sort_values(ascending=False)
```

```python
# Fill missing 'name' using the part before @ in the Email
df['name'] = df.apply(
    lambda row: row['Email'].split('@')[0].replace('.', ' ').replace('_', ' ').title()
    if pd.isnull(row['name']) and pd.notnull(row['Email']) else row['name'],
    axis=1
)
```

```python
# Remove numbers from name column
df['name'] = df['name'].str.replace(r'\d+', '', regex=True)

# Optional: Make names look neat (capitalize each word)
df['name'] = df['name'].str.title().str.strip()
```

```python
df[df['name'].isnull()]
```

```python
df = df[~(df['name'].isnull() & df['Email'].isnull() & df['start_time'].isnull())]
```

```python
df[df['name'].isnull()]
```

```python
# Standardize department_name and batch columns
df['department_name'] = df['department_name'].str.strip().str.title()
df['batch'] = df['batch'].str.strip().str.title()
df['test_status'] = df['test_status'].str.strip().str.title()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[df['candidate_total_score'] > df['total_max_score']]
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['phone'] = df['phone'].fillna('Not Provided')
```

```python
# For students whose test was not started, fill time-related columns with 0
df.loc[df['test_status'].str.lower() == 'not started', ['start_time', 'submit_time', 'test_duration', 'calculated_duration_minutes']] = 0

```

```python
# Fill proctoring_analysis with 0 only where test_status is 'Not Started'
df.loc[df['test_status'].str.lower() == 'not started', 'proctoring_analysis'] = 0
```

```python
# Get all *_score and *_percentage columns
score_cols = [col for col in df.columns if col.endswith('_score') or col.endswith('_percentage')]
```

```python
df['ad_id'] = df['ad_id'].fillna('Not Tracked')
```

```python
# Fill system log columns with 'NA' for missing values
df[['startip', 'endip', 'browser', 'os']] = df[['startip', 'endip', 'browser', 'os']].fillna('NA')
```

```python
# Leave token_paid as NaN or fill with 0 if needed
df['token_paid'] = df['token_paid'].fillna(0)
```

```python
import random

# Step 1: Get existing (non-null) valid pairs of owner and owner_email
owner_pairs = df[['owner', 'owner_email']].dropna().drop_duplicates().values.tolist()

# Step 2: Fill missing values with random valid pairs
for i, row in df.iterrows():
    if pd.isnull(row['owner']) or pd.isnull(row['owner_email']):
        random_owner, random_email = random.choice(owner_pairs)
        df.at[i, 'owner'] = random_owner
        df.at[i, 'owner_email'] = random_email
```

```python
# Fill lead_stage based on test_status
df.loc[(df['lead_stage'].isnull()) & (df['test_status'].str.lower() == 'completed'), 'lead_stage'] = 'Dropped'
df.loc[(df['lead_stage'].isnull()) & (df['test_status'].str.lower() == 'not started'), 'lead_stage'] = 'Not Interested'
```

```python
df['is_token_paid'].value_counts()
df['tabs_switched_flag'].value_counts()
```

```python
# List of all total, score, and percentage columns with 23 missing values
columns_to_fill = [
    'total_max_score', 'candidate_total_score', 'candidate_total_percentage',
    'math_foundation_max_score', 'math_foundation_candidate_score', 'math_foundation_candidate_percentage',
    'data_interpretation_max_score', 'data_interpretation_candidate_score', 'data_interpretation_candidate_percentage',
    'tech_and_computing_fundamentals_max_score', 'tech_and_computing_fundamentals_candidate_score', 'tech_and_computing_fundamentals_candidate_percentage',
    'verbal_reasoning_max_score', 'verbal_reasoning_candidate_score', 'verbal_reasoning_candidate_percentage',
    'logical_reasoning_max_score', 'logical_reasoning_candidate_score', 'logical_reasoning_candidate_percentage',
    'general_reasoning_max_score', 'general_reasoning_candidate_score', 'general_reasoning_candidate_percentage'
]

# Fill with 0 only for rows where test_status is 'Not Started'
df.loc[df['test_status'].str.lower() == 'not started', columns_to_fill] = 0
```

```python
# Fill remaining missing submit_type with 'Manual Submit by typing "End"'
df['submit_type'] = df['submit_type'].fillna('Manual Submit by typing "End"')

```

```python
# Get average proctoring_analysis for each owner
owner_avg = df.groupby('owner')['proctoring_analysis'].mean()

# Fill missing proctoring_analysis with owner average for specific submit types
df['proctoring_analysis'] = df.apply(
    lambda row: owner_avg[row['owner']] if (
        pd.isnull(row['proctoring_analysis']) and
        row['submit_type'] in [
            'Manual Submit by typing "End"',
            'Force Submitted by staff',
            'Auto Submit due to time ended'
        ]
    ) else row['proctoring_analysis'],
    axis=1
)

```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Before Removing Outliers: {col}")
    plt.show()
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower) | (df[column] > upper)]
    print(f"\n📊 {col}")
    print(f"  IQR Range: {lower:.2f} - {upper:.2f}")
    print(f"  Outliers found: {len(outliers)}")

    # Return cleaned dataframe
    return df[(df[column] >= lower) & (df[column] <= upper)]
```

```python
# Loop through all selected columns and remove outliers
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

print("✅ All outliers removed from all selected percentage columns.")
```

```python
def confirm_no_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    print(f"{column}: Outliers Remaining = {len(outliers)}")
```

```python
for col in numeric_cols:
    confirm_no_outliers(df, col)
```
