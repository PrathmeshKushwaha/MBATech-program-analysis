```python
for i in range(len(df_leads['graduation_percentage'])):
    if isinstance(df_leads.loc[i, "graduation_percentage"], str):
        if '%' in df_leads.loc[i, "graduation_percentage"]:
            df_leads.loc[i, "graduation_percentage"] = df_leads.loc[i, "graduation_percentage"].replace('%', '')
        elif df_leads.loc[i, "graduation_percentage"].isdigit():
            df_leads.loc[i, "graduation_percentage"] = int(df_leads.loc[i, "graduation_percentage"])
        else:
            df_leads.loc[i, "graduation_percentage"] = np.nan
```

```python
def convert_to_seconds(x):
    try:
        if pd.isnull(x):
            return np.nan
        # Split mm:ss format
        parts = str(x).split(":")
        if len(parts) == 2:
            minutes = int(parts[0])
            seconds = float(parts[1])
            return minutes * 60 + seconds
        return np.nan
    except:
        return np.nan
```

```python
df_leads['first_call_seconds'] = df_leads['first_call'].apply(convert_to_seconds)
```
