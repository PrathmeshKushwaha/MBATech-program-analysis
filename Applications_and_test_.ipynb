```python
import pandas as pd

# Load both datasets
apps_df = pd.read_csv('/content/Cleaned_Applications (1).csv')
test_df = pd.read_csv('/content/Final_Cleaned_MBA_Tests (1).csv')
```

##Who are serious applicants?

```python
# Strip and lowercase both email columns
apps_df['email'] = apps_df['email'].str.strip().str.lower()
test_df['Email'] = test_df['Email'].str.strip().str.lower()

# Drop rows with missing emails
apps_df = apps_df[apps_df['email'].notnull()]
test_df = test_df[test_df['Email'].notnull()]
```

```python
merged_df = pd.merge(apps_df, test_df, left_on='email', right_on='Email', how='inner')
print("Merged dataset shape:", merged_df.shape)
merged_df.head()
```

```python
print("Number of exact email matches:", merged_df.shape[0])
```

```python
matched_emails = merged_df['email'].unique()
print("Total unique matched emails:", len(matched_emails))
matched_emails[:10]  # preview first 10

```

```python
merged_df.sample(20)
```

```python
print("Total emails in applications:", apps_df['email'].nunique())
print("Total emails in tests:", test_df['Email'].nunique())
print("Matched emails:", len(matched_emails))
```

```python
# Check how many times each email appears
email_counts = merged_df['email'].value_counts()

# Filter to show only duplicates
duplicate_emails = email_counts[email_counts > 1]
print("Number of duplicate emails:", len(duplicate_emails))
duplicate_emails.head()
```

```python
merged_df = merged_df.drop_duplicates(subset='email', keep='first')
```

##Who applied but didn’t attempt the test?

```python
# Clean and standardize emails
apps_df['email'] = apps_df['email'].str.strip().str.lower()
test_df['Email'] = test_df['Email'].str.strip().str.lower()
```

```python
# Emails from application but not in test
not_tested_df = apps_df[~apps_df['email'].isin(test_df['Email'])]

print("Applicants who didn't attempt the test at all:", not_tested_df.shape[0])
not_tested_df[['name', 'email']].head()
```

```python
# Filter from test data
not_started_df = test_df[test_df['test_status'].str.lower() == 'not started']

print("Applicants with test status 'Not Started':", not_started_df.shape[0])
not_started_df[['Email', 'test_status']].head()
```

```python
# Merge "not started" with applications to get full details
not_started_merged = pd.merge(apps_df, not_started_df, left_on='email', right_on='Email', how='inner')

# Combine both
total_not_attempted = pd.concat([not_tested_df, not_started_merged], ignore_index=True)

# Drop duplicates if any
total_not_attempted = total_not_attempted.drop_duplicates(subset='email')

print("Total applicants who didn't attempt the test:", total_not_attempted.shape[0])
total_not_attempted[['name', 'email']].head()
```

##Who are high potential candidates?

```python
strict_high_potential = merged_df[
    (merged_df['overall_academic_score'] >= 75) &
    (merged_df['candidate_total_score'] >= 40) &
    (merged_df['test_status'] == 'Completed')
]
```

```python
print("Number of strict high-potential candidates:", strict_high_potential.shape[0])

strict_high_potential[['email', 'overall_academic_score', 'candidate_total_score', 'test_status','ad_id']].head()

```

##Who to filter out early?

```python
filter_out_early = merged_df[
    (merged_df['overall_academic_score'] < 60) |
    (merged_df['lead_stage_y'].str.lower().isin(['dropped', 'not interested'])) |
    (merged_df['test_status'].str.lower() != 'completed')
]
```

##Which ad campaigns (ad_id) brought in the most high-quality or high-potential candidates?

```python
import pandas as pd
ad_df  = pd.read_csv('/content/Strict_High_Potential.csv')
```

```python
# Group by ad_id and calculate key metrics
ad_id_performance = strict_high_potential.groupby('ad_id').agg({
    'email': 'count',
    'overall_academic_score': 'mean',
    'candidate_total_score': 'mean'
}).rename(columns={
    'email': 'high_potential_count',
    'overall_academic_score': 'avg_academic_score',
    'candidate_total_score': 'avg_test_score'
}).sort_values(by='high_potential_count', ascending=False)
```
