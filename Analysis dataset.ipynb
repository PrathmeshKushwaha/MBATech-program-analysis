Identify Numeric & Categorical Columns


```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of leads per campaign
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='campaign_name', order=df['campaign_name'].value_counts().index)
plt.title("Lead Count per Campaign", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Campaign Name")
plt.ylabel("Number of Leads")
plt.tight_layout()
plt.show()
```

```python
df['total_spent'] = pd.to_numeric(df['total_spent'], errors='coerce')

df.groupby('campaign_name')['total_spent'].sum().plot(kind='bar', color='skyblue')
plt.ylabel('Total Spent')
plt.title('Total Spend per Campaign')
plt.xticks(rotation=45, ha='right')
plt.show()


```

```python
# Ensure CTR is numeric and drop missing values
ctr = pd.to_numeric(df['click_through_rate'], errors='coerce').dropna()

# CTR distribution
plt.figure(figsize=(6, 3))
sns.histplot(ctr, bins=10, color='skyblue', edgecolor='black')
plt.title("CTR Distribution", fontsize=14)
plt.xlabel("Click Through Rate (%)")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the values
platform_counts = df['platform'].value_counts()
print(platform_counts)

# Plot
sns.countplot(data=df, x='platform', order=platform_counts.index)
plt.title("Lead Count by Platform")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_campaigns = df.groupby('campaign_name')['leads'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(x=top_campaigns.values, y=top_campaigns.index)
plt.title("Top 10 Campaigns by Leads")
plt.xlabel("Leads")
plt.ylabel("Campaign Name")
plt.show()
```

```python
# Total Spent vs Clicks
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='total_spent', y='clicks', hue='platform')
plt.title("Total Spent vs Clicks by Platform")
plt.show()
```

```python
# Make sure the columns are numeric
df['impressions'] = pd.to_numeric(df['impressions'], errors='coerce')
df['clicks'] = pd.to_numeric(df['clicks'], errors='coerce')

platform_totals = df.groupby('platform')[['impressions', 'clicks']].sum()

platform_totals.plot(kind='bar', figsize=(8,5), color=['skyblue', 'orange'])
plt.title("Total Impressions vs Clicks by Platform")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Metric")
plt.show()

```
